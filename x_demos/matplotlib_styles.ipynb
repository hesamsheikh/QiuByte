```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

model = LinearRegression()
model.fit(X, y)
X_line = np.array([[0], [2]])
y_line = model.predict(X_line)
```

```python
plt.scatter(X, y, label='Data')
plt.plot(X_line, y_line, color='red', label='Fitted Line')
plt.show()
```

```python
plt.scatter(X, y, label='Data', color='white', edgecolors='k')
plt.plot(X_line, y_line, color='navy', label='Fitted Line')
plt.show()
```

```python
# different Matplotlib styles
styles = plt.style.available

# data and fitted line for each style in separate figures
for i, style in enumerate(styles):
    with plt.style.context(style):
        fig, ax = plt.subplots()
        ax.scatter(X, y, label='Data')
        ax.plot(X_line, y_line, color='red', label='Fitted Line')
        ax.set_title(style)
        ax.legend()
        plt.show()
```
